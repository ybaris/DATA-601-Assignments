```
import csv 
import requests 
import xml.etree.ElementTree as ET

# importing data by reading from the file
tree = ET.parse('/content/xml_containing_html.xml')
root = tree.getroot()
root
```

```
# As an Element, root has a tag and a dictionary of attributes:
print(root.tag)
print(root.attrib)
```

```
# showing tags and attribs
for x in root:
    print(x.tag)
    print(x.attrib)
```

```
# It also has children nodes over which we can iterate:
# parsing XML
for child in root:
    print(child.tag, child.attrib)
    for child_2 in child: 
        print(f"\t{child_2.tag}")
        text = child_2.text.replace("\n", " ")
        print(f"\t\t{text}")
```

```
# Children are nested, and we can access specific child nodes by index:
# first index determines which child will be read.
# second index determines which child_2 will be read. First one is HTML.
root[-1][0].text 
```

```
html_doc = root[2][0].text
html_doc
```

```
from bs4 import BeautifulSoup as BS

# creating a function to count the number of links for each html 
def count_url(html_str):
    soup = BS(html_str, 'html.parser')
    for link in soup.find_all('a'):
        print(link.get('href'))
    number_url = soup.find_all('a')
    print("Number of links: ", len(number_url))
    
count_url(html_doc)
```

```
# iterating the function in a for loop for every html in XML file.

for i in range(125):
    print(root[i].attrib)
    count_url(root[i][0].text)
    print("\n")
```
